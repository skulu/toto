# Monte Carlo Simulation

```python
'''
Types:
6. Ordinary
7. System 7
8. System 8
9. System 9
10. System 10
11. System 11
12. System 12
13. iTOTO
'''

def generate_combi(type):
    '''
    Generates a set of numbers according to bet type.
    Generating winning numbers = type 7 for 6 numbers + 1 additional number
    '''
    if type == 13: type = 12 # iTOTO is System 12 as far as this function is concerned

    tmp = random.randint(1,49)
    combi = [tmp]
    for i in range(type-1): 
        while tmp in combi:
            tmp = random.randint(1,49)
        combi.append(tmp)
    return combi

def check_results(bet, winning_no, type, prize_pool):
    # Split winning numbers to main and additonal numbers (taken as the last number in the 7 generated)
    main = winning_no[0:6]
    additional = winning_no[6]

    # check matches for main numbers
    matches = [0 for i in bet]
    for i, b in enumerate(bet):
        for m in main:
            if b == m:
                matches[i] = 1
                break

    # check match for additional number
    add_match = 0
    for i, b in enumerate(bet):
        if b == additional:
            matches.append(1)
            add_match = 1
            break

    if add_match != 1: 
        matches.append(0)
    
    # Group winning amounts
    grp_amts = {
        1: 0.38*prize_pool,
        2: 0.08*prize_pool,
        3: 0.055*prize_pool,
        4: 0.03*prize_pool,
    }

    # Matrix of possible wins
    win_amt = [
        [0 , grp_amts[1] + grp_amts[2], grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4], grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4] + 1250, grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4] + 3950, grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4] + 8300, grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4] + 14500, (grp_amts[1] + grp_amts[2] + grp_amts[3] + grp_amts[4] + 14500)/28],

        [grp_amts[1], grp_amts[1] + grp_amts[3], grp_amts[1] + grp_amts[3] + 750, grp_amts[1] + grp_amts[3] + 2450, grp_amts[1] + grp_amts[3] + 5300, grp_amts[1] + grp_amts[3] + 9500, grp_amts[1] + grp_amts[3] + 15250, (grp_amts[1] + grp_amts[3] + 15250)/28],

        [grp_amts[2], grp_amts[2] + grp_amts[3] + grp_amts[4], grp_amts[2] + grp_amts[3] + grp_amts[4] + 500, grp_amts[2] + grp_amts[3] + grp_amts[4] + 1600, grp_amts[2] + grp_amts[3] + grp_amts[4] + 3400, grp_amts[2] + grp_amts[3] + grp_amts[4] + 6000, grp_amts[2] + grp_amts[3] + grp_amts[4] + 9500, (grp_amts[3] + grp_amts[4] + 9500)/28],

        [grp_amts[3], grp_amts[3] + 250, grp_amts[3] + 850, grp_amts[3] + 1900, grp_amts[3] + 3500, grp_amts[3] + 5750, grp_amts[3] + 8750, (grp_amts[3] + 8750)/28],

        [grp_amts[4], grp_amts[4] + 150, grp_amts[4] + 490, grp_amts[4] + 1060, grp_amts[4] + 1900, grp_amts[4] + 3050, grp_amts[4] + 4550, (grp_amts[4] + 4550)/28],
        
        [50, 190, 460, 900, 1550, 2450, 3640, 3640/28],

        [25, 85, 190, 350, 575, 875, 1260, 1260/28],
        
        [10, 40, 100, 200, 350, 560, 840, 840/28],
    ]

    # Bet amounts for the different types of bets, 13 is iTOTO
    bet_amt = {
        6: 1,
        7: 7,
        8: 28,
        9: 84,
        10: 210,
        11: 462,
        12: 924,
        13: 33
    }

    # Logic for determining wins
    if sum(matches[0:6]) == 3 and matches[-1] == 0:
        return 7, win_amt[7][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 3 and matches[-1] == 1:
        return 6, win_amt[6][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 4 and matches[-1] == 0:
        return 5, win_amt[5][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 4 and matches[-1] == 1:
        return 4, win_amt[4][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 5 and matches[-1] == 0:
        return 3, win_amt[3][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 5 and matches[-1] == 1:
        return 2, win_amt[2][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 6 and matches[-1] == 0:
        return 1, win_amt[1][type-6] - bet_amt[type]
    elif sum(matches[0:6]) == 6 and matches[-1] == 1:
        return 0, win_amt[0][type-6] - bet_amt[type]
    
    return -1, -bet_amt[type] # no win


def simulate_bet(type, prize_pool):
    bet = generate_combi(type)
    winning_no = generate_combi(7)
    prize_group, win_amt = check_results(bet, winning_no, type, prize_pool)

    return prize_group, win_amt

def run_simulation(times, games_each_time, type, prize_pool):
    '''
    times: number of times to run the simulation
    games_each_time: how many games per simulation
    type: bet type
    prize_pool: suggest to use $3.7M as that is the median prize pool of TOTO
    '''

    results_list =  []
    for i in range(times*games_each_time):
        a = simulate_bet(type, prize_pool)
        results_list.append(list(a))

    result_df = pd.DataFrame(results_list)
    
    # Get net dollar amount at the end of each run of games.
    outcome = []
    for i in range(times):
        # As an example, for 1040 games per simulation, the index goes from [0,1039] then [1040, 2079] etc.
        outcome.append(result_df[games_each_time*i : games_each_time*(i+1)][1].sum())

    print(f'Mean: $ {np.mean(outcome):.2f}')
    print(f'Min: $ {np.min(outcome):.2f}')
    print(f'10th Percentile: $ {np.percentile(outcome, 10):.2f}')
    print(f'25th Percentile: $ {np.percentile(outcome, 25):.2f}')
    print(f'Median: $ {np.median(outcome):.2f}')
    print(f'75th Percentile: $ {np.percentile(outcome, 75):.2f}')
    print(f'90th Percentile: $ {np.percentile(outcome, 90):.2f}')
    print(f'Max: $ {np.max(outcome):.2f}')

    return result_df, outcome
```
